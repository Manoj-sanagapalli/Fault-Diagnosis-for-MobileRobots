```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.utils import resample
import datetime as dt
```

```python
file = '/home/roscon/Desktop/Data_latest/Model/Trail1/0.xlsx'

file2 = '/home/roscon/Desktop/Data_latest/Model/Trail1/1.xlsx'
data= pd.read_excel(file,sheet_name=0)
data2 = pd.read_excel(file2,sheet_name=0)
print(data2.shape)
data = data.iloc[24:]
data2 = data2.iloc[20:]
time = (data2.index)*0.1
#current
right_current = data["Right current"]
right_current2 = data2["Right current"]

left_current = data["Left current"]
left_current2 = data2["Left current"]

fig = plt.figure(figsize = (12,4))

#ax1 = fig.add_subplot(3, 2, 1)
ax1 = fig.add_subplot(1, 2, 1)
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('Left current (*100 mA)')

ax1.plot(time, left_current)
ax1.plot(time, left_current2)


plt.title("Left Current")
plt.legend(["No obstacle", "Obstacle"])

#ax2 = fig.add_subplot(3, 2, 2)
ax2 = fig.add_subplot(1, 2, 2)
ax2.set_xlabel('Time (seconds)')
ax2.set_ylabel('Right current (*100 mA)')
ax2.plot(time, right_current)
ax2.plot(time, right_current2)

plt.title("Right Current")
plt.legend(["No obstacle", "Obstacle"])
```
