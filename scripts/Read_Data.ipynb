```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.utils import resample
import datetime as dt
#%matplotlib notebook

def plot(dataframe,title):
    time = (dataframe.index)*0.1
    actual_vel = dataframe["actual_vel"]
    right_current = dataframe["Right current"]
    left_current = dataframe["Left current"]
    bumper = dataframe["Bumper"]
    wheel = dataframe["Wheeldrop"]
    fig = plt.figure(figsize = (12,7))

    color1 = 'tab:orange'
    color2 = 'tab:green'

    ax1 = fig.add_subplot(2,2,1)
    ax1.set_xlabel('Time (seconds)')
    ax1.set_ylabel('Current (*100 mA)')
    ax1.plot(time, right_current,color=color1,label='Right current')
    ax1.legend(loc="lower center")
    ax1.grid()

    
    ##############################################################################################################

    ax3 = fig.add_subplot(2,2,2)
    ax3.set_xlabel('Time (seconds)')
    ax3.set_ylabel('Current (*100 mA)')
    ax3.plot(time, left_current,color = color2,label='Left current')
    ax3.legend(loc="lower left")
    ax3.grid()

    

    ##############################################################################################################
    
    ax5 = fig.add_subplot(2,2,3)
    ax5.set_xlabel('Time (seconds)')
    ax5.set_ylabel('Actual velocity (m/sec)')
    ax5.plot(time, actual_vel, label = 'Actual Velocity')
    ax5.grid()
   
    ax6 = fig.add_subplot(2,2,4)
    ax6.set_xlabel('Time (seconds)')
    ax6.set_ylabel('Bumper')
    ax6.plot(time, bumper,color = color2,label='Bumper')
    ax6.legend(loc="lower left")
    ax6.grid()

    ax7 = ax6.twinx()
    ax7.set_ylabel('Wheel drop')
    ax7.plot(time, wheel,label='Wheel drop')
    ax7.legend(loc="lower right")





    
    
    
    

    fig.tight_layout(pad=4.0)

    plt.suptitle(title)
    plt.show()
```
